```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
data=requests.get("https://yemenshabab.net/ar/sections/NEWS/Political")
```

```python
data.text
```

```python
soup=BeautifulSoup(data.text,'html.parser')
```

```python
titel=soup.select('.pt-4')
```

```python
len(titel)
```

```python
news=[]
for i,v in enumerate(titel):
    print(i+1,"-",v.text)
    news.append(v.text)
```

```python
news
```

```python
df=pd.DataFrame()
```

```python
df["titelNews"]=news
```

```python
df
```

```python
df.to_csv("news.csv",index=False)
```
