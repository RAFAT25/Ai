```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res= requests.get("https://www.aljazeera.net/")
```

```python
res
```

```python
res.text
```

```python
soup=BeautifulSoup(res.text,'html.parser')
```

```python
title=soup.select('.article-card__title')
```

```python
len(title)
```

```python
news=[]
for i,v in enumerate(title):
    print(i+1,"-",v.text)
    news.append(v.text)
```

```python
news
```

```python
df=pd.DataFrame()
```

```python
df["titelNews"]=news
```

```python
df
```

```python
df.to_csv("news.csv",index=False)
```
